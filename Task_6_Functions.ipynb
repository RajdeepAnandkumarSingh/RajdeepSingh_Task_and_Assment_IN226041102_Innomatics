```python
def manage_parking(capacity, logs):
    current = 0
    peak = 0
    for action in logs:
        current = max(0, current + 1 if action == "IN" else current - 1)
        peak = max(peak, current)

    status = "FULL - Alert!" if current >= capacity else "Available"
    return current, status

# Test
count, stat = manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
print(f"Currently Parked: {count}\nStatus: {stat}")
```

```python
def calculate_eta(dist, traffic, weather):
    multipliers = {"Low": 1.0, "Medium": 1.5, "High": 2.2}
    weather_delays = {"Rainy": 15, "Snowy": 25, "Stormy": 40}

    base_time = dist * 2
    traffic_time = base_time * multipliers.get(traffic, 1.0)
    final_eta = round(traffic_time + weather_delays.get(weather, 0))

    return final_eta

# Test
print(f"Estimated Delivery Time: {calculate_eta(8, 'High', 'Rainy')} minutes")
```

```python
def analyze_bookings(total, booked_list):
    percent = (len(booked_list) / total) * 100

    if percent == 100: status = "Housefull"
    elif percent >= 80: status = "Almost Full"
    else: status = "Seats Available"

    suggestion = "Open new show!" if percent > 85 else "No action"

    return percent, status, suggestion

# Test
occ, stat, sug = analyze_bookings(200, [1]*150)
print(f"Occupancy: {occ}%\nStatus: {stat}\nAction: {sug}")
```

```python
def classify_server_load(readings):
    avg = sum(readings) / len(readings)

    if avg >= 80:
        status = "Critical"
    elif 50 <= avg < 80:
        status = "Warning"
    else:
        status = "Normal"

    return round(avg), status

# Test
readings = [45, 60, 70, 85, 90]
avg_load, status = classify_server_load(readings)

print(f"Average CPU Load: {avg_load}%")
print(f"Server Status: {status}")
```

```python
def check_resource_usage(usage_dict, threshold=8):
    overused = [res for res, hours in usage_dict.items() if hours > threshold]

    alert = "Yes" if overused else "No"

    return overused, alert

# Test
usage = {"Projector": 6, "AC": 9, "Lights": 4}
over, alert_status = check_resource_usage(usage)

print(f"Overused Resources: {', '.join(over) if over else 'None'}")
print(f"Energy Alert: {alert_status}")
```

```python
def manage_registrations(capacity, total_signups):
    confirmed = min(total_signups, capacity)
    waitlisted = max(0, total_signups - capacity)

    status = "Closed" if total_signups >= capacity else "Open"

    return confirmed, waitlisted, status

# Test
cap = 100
signups = 105

conf, wait, stat = manage_registrations(cap, signups)

print(f"Confirmed Registrations: {conf}")
print(f"Waitlisted Users: {wait}")
print(f"Registration Status: {stat}")
```
